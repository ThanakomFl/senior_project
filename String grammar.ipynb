```python
import cv2
import matplotlib.pyplot as plt
import os
import imutils
import numpy as np
from skimage.feature import hog
import pdb
```

```python
all_datalist = os.listdir('./dataset/')
mouth_cascade = cv2.CascadeClassifier('./src/mouth.xml')
ds_factor = 0.5
```

# Find average

```python
fix_size_w = 150
fix_size_h = 75
dim = (fix_size_w, fix_size_h)
```

```python
avg_f = []
crop = [450,100,550,250]
for idx in range(len(all_datalist)):
    get_data = os.listdir('./dataset/'+all_datalist[idx])
    for i in range(0,len(get_data)-4):
        cap = cv2.VideoCapture('./dataset/'+all_datalist[idx]+'/'+get_data[i])
        haveFrame,im = cap.read()
        avg_f_i = np.zeros((fix_size_h,fix_size_w))
        j = 0
        while(cap.isOpened()):
            print("pp : "+str(idx)+" Video : "+str(i)+" Frame : "+str(j))
            haveFrame,im = cap.read()
            if (not haveFrame) or (cv2.waitKey(1) & 0xFF == ord('q') or j > 50):
                break
            im = np.rot90(im)
            im = np.flipud(im)
            im = cv2.resize(im, None, fx=ds_factor, fy=ds_factor, interpolation=cv2.INTER_AREA)
            gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
            mouth_rects = mouth_cascade.detectMultiScale(gray, 1.7, 11)
            for (x,y,w,h) in mouth_rects:
                y = int(y - 0.15*h)
                if(h > 100 and y > 400):
                    im = im[y:y+w,x:x+h]
                    break
            if(len(im) > 200):
                continue
            im = cv2.resize(im, dim, interpolation = cv2.INTER_AREA)
#             cv2.imwrite('./output/'+str(i)+"_"+str(j)+".jpg",im)
            im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
            avg_f_i = avg_f_i + im
#             cv2.imshow('im', im)
#             cv2.moveWindow('im', 800, 50)
            j = j +1
        avg_f.append(avg_f_i/j) 
        cap.release()
        cv2.destroyAllWindows()
```

```python
avg_f_ii = np.zeros((len(avg_f[0]),len(avg_f[0][0])))
for i in range(len(avg_f)):
    avg_f_ii = avg_f_ii + avg_f[i]
avg_f_ii = avg_f_ii / len(avg_f)
```

# Find Dif per Frame

```python
true_label = []
stgm = []
for idx in range(len(all_datalist)):
    get_data = os.listdir('./dataset/'+all_datalist[idx])
    for i in range(5,len(get_data)):
        cap = cv2.VideoCapture('./dataset/'+all_datalist[idx]+'/'+get_data[i])
        haveFrame,im = cap.read()
        j = 0
        avg_f_i = np.zeros((fix_size_h,fix_size_w))  
        stgm_i = []
        true_label_i = []
        while(cap.isOpened()):
            haveFrame,im = cap.read()
            print("pp : "+str(idx)+" Video : "+str(i)+" Frame : "+str(j))
            if (not haveFrame) or (cv2.waitKey(1) & 0xFF == ord('q') or j > 10):
                break
            im = np.rot90(im)
            im = np.flipud(im)
            im = cv2.resize(im, None, fx=ds_factor, fy=ds_factor, interpolation=cv2.INTER_AREA)
            gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
            mouth_rects = mouth_cascade.detectMultiScale(gray, 1.7, 11)
            for (x,y,w,h) in mouth_rects:
                y = int(y - 0.15*h)
                if(h > 100 and y > 400):
                    im = im[y:y+w,x:x+h]
                    break
            if(len(im) > 200):
                continue
            im = cv2.resize(im, dim, interpolation = cv2.INTER_AREA)
            im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)    
            Dif_f = im-avg_f_ii
            Blur_f = cv2.GaussianBlur(im,(3,3),cv2.BORDER_DEFAULT,1)
            Fi_f = Dif_f / Blur_f
            fd, hog_image = hog(Fi_f, orientations=16, pixels_per_cell=(2,2),
                            cells_per_block=(2,2),visualize=True)
            stgm.append(fd)
            cv2.imshow('im_dif', Dif_f)
            cv2.moveWindow('im_dif', 800, 50)
            cv2.imshow('im_blur', Blur_f)
            cv2.moveWindow('im_blur', 1150,50)
            cv2.imshow('Fi_f', Fi_f)
            cv2.moveWindow('Fi_f', 450,50)
            cv2.imshow('Hog_img', hog_image)
            cv2.moveWindow('Hog_img', 100,50)
            cv2.imwrite('./output/'+str(i)+"_"+str(j)+".jpg",Fi_f*255)
            true_label.append(idx)
            j = j+1
        print('Get hog for 1 video')
        cap.release()
        cv2.destroyAllWindows()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(stgm,true_label, test_size=0.33, random_state=42)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train) 
```

```python
y_pred = knn.predict(X_test)
accuracy_score(y_test, y_pred)
```
